Problem 1 is an exercise in statistics and probabilities using python. 

1st thing: N pairs of random variables x and y from uniform pdf between 0 and 1

2nd interpreted as coordinates in the X,Y plane. Pairs will populate uniformly a square of area = 1. 

3rd the number of n points that lie within the semicircle of radius 1 will vary from trial to trial, and will follow a binomial distribution (eq on sheet).

f is the ratio of the area of the semicircle to that of the square: f = $\frac{1}{4}\pi$.

Estimate: f as $\tilde{f} = \frac{n}{N}$ and compute $\pi$ as $\tilde{\pi} = 4\pi$

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
```

```python
s = np.random.uniform(0,1,1000)
count, bins, ignored = plt.hist(s, 15, density=True)
plt.plot(bins, np.ones_like(bins), linewidth=2, color ='r')
plt.show()
```

```python
W_values = [1.0, 2.0, 3.0]
linestyles = ['-', '--', ':']
mu = 0
x = np.linspace(-2, 2, 1000)

#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(5, 3.75))

for W, ls in zip(W_values, linestyles):
    left = mu - 0.5 * W
    dist = uniform(left, W)

    plt.plot(x, dist.pdf(x), ls=ls, c='black',
             label=r'$\mu=%i,\ W=%i$' % (mu, W))

plt.xlim(-1.7, 1.7)
plt.ylim(0, 1.2)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu, W)$')
plt.title('Uniform Distribution')

plt.legend()
plt.show()
```
